```python
import os

import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_results(ax, name):

    results = np.load(os.path.join("./results", name+".npy"), allow_pickle=True).item()

    ax[0].plot(results["w"], label=name)
    ax[1].plot(results["loss"], label=name)
    ax[2].plot(results["acc"], label=name)
```

Greedy, unconstrained

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

plot_results(ax, "SGD")
plot_results(ax, "IMT")
plot_results(ax, "LAST")
plot_results(ax, "DHT")
plot_results(ax, "Joint")

ax[0].legend()
ax[0].set_xlabel("Iteration")
ax[0].set_ylabel(f"Distance between $w^t$ and $w^*$")

ax[1].legend()
ax[1].set_xlabel("Iteration")
ax[1].set_ylabel("Training Loss")

ax[2].legend()
ax[2].set_xlabel("Iteration")
ax[2].set_ylabel("Test Accuracy")
```

Greedy, constrained

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

plot_results(ax, "SGD")
plot_results(ax, "IMT")
plot_results(ax, "LAST_soft")
plot_results(ax, "DHT_xrange")
plot_results(ax, "Joint_xrange_soft")

ax[0].legend()
ax[0].set_xlabel("Iteration")
ax[0].set_ylabel(f"Distance between $w^t$ and $w^*$")

ax[1].legend()
ax[1].set_xlabel("Iteration")
ax[1].set_ylabel("Training Loss")

ax[2].legend()
ax[2].set_xlabel("Iteration")
ax[2].set_ylabel("Test Accuracy")
```

Greedy vs Parametrised DHT

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 5))

plot_results(ax, "SGD")
plot_results(ax, "IMT")
plot_results(ax, "DHT")
plot_results(ax, "Parametrised")

ax[0].legend()
ax[0].set_xlabel("Iteration")
ax[0].set_ylabel(f"Distance between $w^t$ and $w^*$")

ax[1].legend()
ax[1].set_xlabel("Iteration")
ax[1].set_ylabel("Training Loss")

ax[2].legend()
ax[2].set_xlabel("Iteration")
ax[2].set_ylabel("Test Accuracy")
```
